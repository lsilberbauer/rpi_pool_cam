```python
%matplotlib inline
#The line above is necesary to show Matplotlib's plots inside a Jupyter Notebook

import sys
sys.path.append('../')
import yaml
import cv2 
from matplotlib import pyplot as plt
import numpy as np
import math
from src.cam_image_processor import CamImageProcessor
import os

directory = "../data"

with open('../config.yaml', 'r') as file:
    config = yaml.safe_load(file)
    
for image_file in os.listdir(directory):
    if image_file.endswith(".jpg"):
        image = cv2.imread(os.path.join(directory, image_file))

        myCIP = CamImageProcessor(config, image)

        print(f"File {image_file}")
        print(myCIP.get_led_json())

        plt.imshow(cv2.cvtColor(myCIP.get_led_annotations(), cv2.COLOR_BGR2RGB))
        plt.show()
```
